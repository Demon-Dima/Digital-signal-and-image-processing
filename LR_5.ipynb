```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from skimage import io
```


Завдання 1

```python
img_link = "https://raw.githubusercontent.com/mosolab/DIP/master/examples/1_8.png"
img = io.imread(img_link)
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(img, cmap='gray')
plt.show();
```

```python
bin_img = img.copy()
threshold = 10
max_val = 255
min_val = 0
```

```python
for row in range(img.shape[0]):
    for col in range(img.shape[1]):
        if (img[row][col]) > threshold:
            bin_img[row][col] = max_val
        else:
            bin_img[row][col] = min_val
```

```python
plt.figure(figsize=(6, 6))
plt.imshow(bin_img, cmap='gray')
plt.show();
```

```python
kernel = np.ones((2,2),np.uint8)
```

```python
erosion = cv2.erode(bin_img,kernel)
dilation = cv2.dilate(bin_img,kernel)
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(1,2,1)
plt.imshow(erosion, cmap='gray')
plt.title("Erosion")

plt.subplot(1,2,2)
plt.imshow(dilation, cmap='gray')
plt.title("Dilation")


plt.show();
```

Завдання 2

```python
img_link = "https://raw.githubusercontent.com/mosolab/DIP/master/examples/street.jpg"
img = io.imread(img_link)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)   
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(img, cmap='gray')
plt.show();
```

```python
bin_img = img.copy()
threshold = 170
max_val = 255
min_val = 0
```

```python
for row in range(img.shape[0]):
    for col in range(img.shape[1]):
        if (img[row][col]) > threshold:
            bin_img[row][col] = max_val
        else:
            bin_img[row][col] = min_val
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(bin_img, cmap='gray')
plt.show();
```

```python
kernel = np.ones((2,2),np.uint8)
```

```python
erosion = cv2.erode(bin_img,kernel)
```

```python
plt.figure(figsize=(15, 15))
plt.imshow(erosion, cmap='gray')
plt.show();
```

```python
opening = cv2.morphologyEx(erosion , cv2.MORPH_OPEN, kernel)
```

```python

plt.figure(figsize=(10, 10))
plt.imshow(opening, cmap='gray')
plt.show();
```

```python
plt.figure(figsize=(25, 20))

plt.subplot(1,2,1)
plt.imshow(erosion, cmap='gray')
plt.title("Erosion")

plt.subplot(1,2,2)
plt.imshow(opening, cmap='gray')
plt.title("Opening")


plt.show();
```
